```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import cos, sin, pi
```

```python
def Cos(x,L,n):
  return np.cos(2*np.pi*n*x/L)

def Sin(x,L,n):
  return np.sin(2*np.pi*n*x/L)
```

# Comparing Function with Different Periodicity

* In the cell below we plot the functions
$$
\cos\left(\frac{2\pi x}{3}\right) \quad \cos\left(\frac{2\pi x}{5.6}\right)
$$
to see the effect of changing the periodicity $L$.

```python
plt.figure(figsize=(12,9))

x=np.linspace(-5.6,5.6,501)

plt.plot(x,Cos(x,3,1),label='$\\cos\\left(\\frac{2\pi x}{3}\\right)$',color='xkcd:orange red')
plt.plot(x,Cos(x,5.6,1),label='$\\cos\\left(\\frac{2\pi x}{5.6}\\right)$',color='xkcd:dull blue')

plt.xticks([-5.6,-3,0,3,5.6])


plt.xlabel('$x$')
plt.ylabel('$\\cos\\left(\\frac{2\pi x}{L}\\right)$')
plt.legend(ncol=2,loc='upper left', bbox_to_anchor=(-0.,1.1));
```

# An $L$-periodic Function

* In the cells below, we plot the 2-periodic function

4 \text{ if }0\leq x < 1.
\end{cases}
$$

```python
plt.figure(figsize=(8,5))

plt.plot([-1,0,0,1,1,2,2,3,3],[0,0,4,4,0,0,4,4,0],color='xkcd:bright blue')

plt.xlabel('$x$')
plt.ylabel('$f(x)$')
```

## Step 1

* In the code cells below we define python functions for the Fourier coefficients $a_{0}$, $a_{n}$ and $b_{n}$.


* The coefficients defined below are those found in __Lecture 11 - Example 4__ i.e.

$$
a_{0}=2\quad a_{n}=0\quad b_{n}=\frac{4}{n\pi}(1-\cos(n\pi)).
$$

```python
a0=2
```

```python
def a(n):
    return 0
```

```python
def b(n):
    return (4/n/pi)*(1-cos(n*pi))
```

## Step 2

* We now define a python function to compute the _partial sum_ of the first $N$ terms of the Fourier series i.e.

$$
f(x,N) = a_{0} +\sum_{n=1}^{N}a_{n}\cos\left(\frac{2\pi n x}{L}\right) + \sum_{n=1}^{N}b_{n}\sin\left(\frac{2\pi n x}{L}\right)
$$

```python
def f(x,N,L):
    return a0 + sum(a(n)*cos(2*pi*n*x/L) + b(n)*sin(2*pi*n*x/L) for n in range(1,N+1))
```

## Step 3

* Define an interval for $x$, we will use $0\leq x\leq 2L$ with 501 point, and then plot $f(x,N)$.


* We are going to plot the partial sums of the first __5__, __50__ and __500__ terms.

```python
L=2
X=np.linspace(-1,-1+2*L,501)
```

```python
N=5

plt.figure(figsize=(8,5))
plt.plot([-1,0,0,1,1,2,2,3,3],[0,0,4,4,0,0,4,4,0],color='darkslategray',alpha=0.5,label='$f(x)$')
plt.plot(X,f(X,N,L),label='$N=%s$ terms' % N,color='xkcd:bright blue')

plt.xlabel('$x$')
plt.ylabel('Fourier Sum')
plt.legend(ncol=2,loc='upper left', bbox_to_anchor=(-0.,1.1));
```

```python
N=50

plt.figure(figsize=(8,5))
plt.plot([-1,0,0,1,1,2,2,3,3],[0,0,4,4,0,0,4,4,0],color='darkslategray',alpha=0.5,label='$f(x)$')
plt.plot(X,f(X,N,L),label='$N=%s$ terms' % N,color='xkcd:bright blue')

plt.xlabel('$x$')
plt.ylabel('Fourier Sum')
plt.legend(ncol=2,loc='upper left', bbox_to_anchor=(-0.,1.1));
```

```python
N=500

plt.figure(figsize=(8,5))
plt.plot([-1,0,0,1,1,2,2,3,3],[0,0,4,4,0,0,4,4,0],color='darkslategray',alpha=0.5,label='$f(x)$')
plt.plot(X,f(X,N,L),label='$N=%s$ terms' % N,color='xkcd:bright blue')

plt.xlabel('$x$')
plt.ylabel('Fourier Sum')
plt.legend(ncol=2,loc='upper left', bbox_to_anchor=(-0.,1.1));
```
